# Credit Card Default Prediction

## Internship Project

### Objective
Build a Machine Learning model to predict whether a customer is likely to default on credit payments.

### Dataset Features
- Income
- Age
- Loan Amount
- Loan to Income Ratio
- Default (Target Variable)


```python
# Import required libraries
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

```python
import pandas as pd

# Load dataset
!unzip -o '/content/Credit_Card_Default_Prediction_Project (2).zip'
data = pd.read_csv('credit_default_data.csv')
data.head()
```

```python
# Dataset information
data.info()
```

```python
# Statistical summary
data.describe()
```

```python
# Checking default distribution
data['Default'].value_counts()
```

```python
# Define features and target

X = data.drop('Default', axis=1)
y = data['Default']
```

```python
# Split data into training and testing sets

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42
)
```

```python
# Train Logistic Regression model

model = LogisticRegression()
model.fit(X_train, y_train)
```

```python
# Prediction

y_pred = model.predict(X_test)
y_pred
```

```python
# Model evaluation

print("Accuracy:", accuracy_score(y_test, y_pred))
print("\nConfusion Matrix:")
print(confusion_matrix(y_test, y_pred))

print("\nClassification Report:")
print(classification_report(y_test, y_pred))
```

```python
# Visualization

plt.scatter(data['Income'], data['Loan'], c=data['Default'])
plt.xlabel('Income')
plt.ylabel('Loan Amount')
plt.title('Credit Default Pattern')
plt.show()
```

## Conclusion

The Logistic Regression model can classify customers based on financial factors and identify potential credit default risks.
